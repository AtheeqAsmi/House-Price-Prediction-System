The market analysis domain, particularly within the context of house price prediction, holds significant potential and has proven to be highly beneficial and widely applicable across the industry. According to the work of Thamarai and Malarvizhi (2020), this decade has seen a remarkable increase in both the scope and sophistication of the tools and techniques used in this area. The continuous advancements in machine learning models and algorithms have significantly improved the ability to predict house prices with higher accuracy. Each year, the value of houses tends to rise, and this creates a growing need for AI-driven models that can effectively assist buyers and investors in predicting house prices based on various factors tailored to their specific needs and preferences. The ongoing improvements in predictive models, as highlighted by Thamarai and Malarvizhi (2020), further reinforce the importance of using advanced technology to make data-driven predictions that can enhance decision-making processes in real estate. With the increasing demand for real estate, especially in urban areas, these AI models offer a crucial tool in helping users navigate the complexities of the housing market and make informed investment choices.

Many machine learning models such as linear regression, artificial neural networks and more, are used to predict house prices. According to Zulkifley et al.(2020) the house price prediction model has numerous advantages as it can provide all necessary information to home builders, investors, home buyers and more, including the valuation of the present market rates, which help in setting the home prices. This model can help such clients to choose the perfect property depending on their budget as well(Zulkifley et al., 2020).

According to Zulkifley et al.(2020), there are two ways to predict the prices of houses. Firstly by focusing on the characteristics of the house and secondly by focusing on the model used for the prediction. A research done by Rahadi et al.(2016) used a conceptual model and questionnaires to analyze the current house prices in Jakarta. Based on the results of the research, the factors that determine the price of the house differ for each, hence proving the validity of this analysis(Zulkifley et al., 2020)

Three factors are what affect the price of a house according to Nur et al.(2017): The location, neighborhood and structural condition of the house.

1) **Location**:

  Osmadi et al.(2015) in his study saw that the prices of houses significantly depend on the location of it. All of these studies state a very strong connection between the location attributes and the price that determines the house, such as distace from the nearest hospital, distance from the nearest suermarket and so on. (Osmandi et al. 2015)

2) **Neighborhood**:

  According to Chin and Chau(2003), the community social status, distance to shopping malls and efficiency of the public education around that area tend to have an impact on the house prices.(Chin and Chau, 2003)

3) **Structure**:

  The structure of a house is typically the number of bedrooms, number of bathrooms, the sqarefeet area of the property, number of floors and many more. Owusu-Manu et al.(2019) states that all these structural attributes have a very strong relationship to the rising house prices.

In order to add on these insights, the rapid rise in AI and machine learning have completely changed how price predictions are made. AI driven models like random forest gradient boosting machines, neural networks, linear regression and so on, These models can handle intricate non-linear interactions between features including location, neighborhood quality, and structural qualities, providing predictions that are more precise and scalable (Chen et al., 2019).

Handling the outliers

The method used to handle the outliers is IQR (Inter Quartile Range). This basically segregates the data into quartiles, and the lowerbound and upper bound are calculated. Any value below the lower bound or any value above the upper bound is considered an outlier.

```python
Q1 = data["price"].quantile(0.25)
Q3 = data["price"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = data[(data["price"] < lower_bound) | (data["price"] > upper_bound)]
print(f"the lower bound is: {lower_bound}")
print(f"the upper bound is: {upper_bound}")
```

Now let us visuallize the prices after identifying the outliers, and removing them

```python
plt.figure(figsize=(10, 8))
filtered_data = data[(data["price"] >= lower_bound) & (data["price"] <= upper_bound)]["price"]
plt.hist(filtered_data, bins=50, color='green', label="Price Distribution")
plt.title("Distribution of House Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.legend()
plt.gca().xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
```

The graph is much less skewed than it was before handling the outliers as you can see

Now let us create a new data frame, without the outliers, that will be used to train the model later

```python
filtered_data_frame = data[(data["price"] >= lower_bound) & (data["price"] <= upper_bound)]
print(filtered_data_frame.head())
removed_records = data.shape[0] - filtered_data_frame.shape[0]
print(f"\nNumber of removed records: {removed_records}")
```

convert the city categorical variables into numerical representation. This is because machine learning models typically work best with numeric data

```python
filtered_data_frame = pd.get_dummies(filtered_data_frame, columns=['city','statezip'], prefix=['city','statezip'])
filtered_data_frame.head()
```

```python
filtered_data_frame.shape
```

What's done above is, each city has been made as a column, and if a house is in that respective city, the value in that respective city will have a value of 'True(1)', while 'False(0)' for the rest.
The same is applied for statezip.

splitting the target variable from the rest of the data. (splitting into x and y)

```python
x = filtered_data_frame.drop(columns=['price']) #features that help to predict the target variable
y = filtered_data_frame['price'] #target variable 'price' which is to be predicted
```

```python
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

scaling the features. As some of the features in the dataset consists of much larger values than some other features, there is an incosistency. Scaling the features ensure that all features are in the same scale that prevent some larger values dominating over the smaller values

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

model and prediction

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

Calculating the Mean Absolute Error(MAE) and the R-squared(R2), to calculate how far the predicted results are from the actual values, and the variance of the model
